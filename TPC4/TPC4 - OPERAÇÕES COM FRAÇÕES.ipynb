## Resumo

O seguinte trabalho tem como objetivo crirar uma aplicação capaz de realizar operações com frações (somar, subtrair, dividir e multiplicar). Os resultados de todas as operações têm ainda de estar na forma irredutível.

Para representar funções em Python decidi usar listas de modo a que o primeiro valor fosse o numerador e o segundo o denominador.

Comecei por criar a função __fração(numerador, denominador)__ que recebe dois valores e coloca-os numa lista. De seguida crei as funções __soma(f1, f2)__, __multiplicar(f1, f2)__ e __subtrair(f1, f2)__ que realiza as quatro operações (soma, subtração, divisão e multiplicação). Para efetuar a divisão usamos a função __multiplicar(f1, f2)__ e na segunda fração apenas invertemos a ordem do numerador com o denominador.

Por fim, para colocar os resultados das operações na forma irredutível crei a função __MDC(resultado)__ que calcula o máximo divisor comum do numerador e denominador e faz _return_ da divisão de cada um destes pelo o mdc.

## Algoritmo

```python
def menu():
    print("""
Operações das Funções:
(1) Fração na forma irredutível.
(2) Somar frações.
(3) Subtrair frações.
(4) Multiplicar frações.
(5) Dividir frações.
(6) Encerrar o programa.""")

def fração(numerador, denominador):
    return [numerador, denominador]

def MDC(resultado):
    f=resultado.copy()
    while(f[1] != 0):
        resto = f[0] % f[1]
        f[0] = f[1]
        f[1] = resto
  
    return print("O resultado da operação foi", str(int(resultado[0]/f[0]))+"/"+str(int(resultado[1]/f[0])),".")

def soma(f1, f2):
    resultado=[f1[0]*f2[1]+f2[0]*f1[1], f1[1]*f2[1]]
    return MDC(resultado)
    
def multiplicar(f1, f2):
    resultado=[f1[0]*f2[0], f1[1]*f2[1]]
    return MDC(resultado)

def subtrair(f1, f2):
    resultado=[f1[0]*f2[1]-f2[0]*f1[1], f1[1]*f2[1]]
    return MDC(resultado)

variável="1"
while variável!="0":
    menu()
    opção_utilizador=input("Qual a sua opção: ")
    
    if opção_utilizador=="6":
        variável="0"
        print("O programa encerrou. Obrigado e até à próxima.")
    
    elif opção_utilizador=="1":
        numerador=int(input("Introduza o valor do numerador: "))
        denominador=int(input("Introduza o valor do denominador: "))
        resultado=fração(numerador, denominador)
        MDC(resultado)
    
    elif opção_utilizador=="2":
        numerador_1=int(input("Introduza o valor do 1.º numerador: "))
        denominador_1=int(input("Introduza o valor do 1.º denominador: "))
        f1=fração(numerador_1, denominador_1)
        numerador_2=int(input("Introduza o valor do 2.º numerador: "))
        denominador_2=int(input("Introduza o valor do 2.º denominador: "))
        f2=fração(numerador_2, denominador_2)
        soma(f1, f2)
        
    elif opção_utilizador=="3":
        numerador_1=int(input("Introduza o valor do 1.º numerador: "))
        denominador_1=int(input("Introduza o valor do 1.º denominador: "))
        f1=fração(numerador_1, denominador_1)
        numerador_2=int(input("Introduza o valor do 2.º numerador: "))
        denominador_2=int(input("Introduza o valor do 2.º denominador: "))
        f2=fração(numerador_2, denominador_2)
        subtrair(f1, f2)
        
    elif opção_utilizador=="4":
        numerador_1=int(input("Introduza o valor do 1.º numerador: "))
        denominador_1=int(input("Introduza o valor do 1.º denominador: "))
        f1=fração(numerador_1, denominador_1)
        numerador_2=int(input("Introduza o valor do 2.º numerador: "))
        denominador_2=int(input("Introduza o valor do 2.º denominador: "))
        f2=fração(numerador_2, denominador_2)
        multiplicar(f1, f2)
        
    elif opção_utilizador=="5":
        numerador_1=int(input("Introduza o valor do 1.º numerador: "))
        denominador_1=int(input("Introduza o valor do 1.º denominador: "))
        f1=fração(numerador_1, denominador_1)
        numerador_2=int(input("Introduza o valor do 2.º numerador: "))
        denominador_2=int(input("Introduza o valor do 2.º denominador: "))
        f2=fração(denominador_2, numerador_2)
        multiplicar(f1, f2)
    
    else:
        print("A opção selecionada não é válida.")
```
